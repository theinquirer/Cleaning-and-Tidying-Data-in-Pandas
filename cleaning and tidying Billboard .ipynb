```python
import pandas as pd 
```

```python
billboard =pd.read_csv(r'C:/Users/amash/OneDrive/Desktop/cleaningData/billboard.csv')

billboard
```

```python
#columns that remian unchanged
billboard_melt = pd.melt(
    billboard,
    id_vars =['year', 'artist','track','date.entered', 'time', ],
    var_name='week',
    value_name ='rating')

billboard_melt
```

```python
billboard_melt.head()
```

```python
#  create a song data set
billboard_songs = billboard[['year', 'artist', 'track', 'time']]
billboard_songs
```

```python
# create a column id using a range-a number generator 
billboard_songs['id'] = range(len(billboard_songs))
```

```python
billboard_songs.head()
```

```python
#save dataframe to a csv file
billboard_songs.to_csv('billboard_songs.csv', index =False)
```

```python
#Merge the billboard melt data set to the billboard song dataset 
billboard_ratings = billboard_melt.merge(
     billboard_songs,  on =['year','artist','track', 'time'])
```

```python
billboard_ratings.head()
```

```python
billboard_ratings = billboard_ratings[['id','date.entered','week','rating']]
```
